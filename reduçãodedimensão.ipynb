```python
from PIL import Image

def converter_imagem(nome_arquivo):
    # Abrir a imagem
    imagem_colorida = Image.open(nome_arquivo)

    # Converter para escala de cinza
    imagem_cinza = imagem_colorida.convert('L')

    # Salvar a imagem em escala de cinza
    nome_arquivo_cinza = nome_arquivo.replace('.jpg', 'cinza.jpg')
    imagem_cinza.save(nome_arquivo_cinza)

    # Converter para preto e branco
    limiar = 127
    imagem_pb = imagem_cinza.point(lambda p: p > limiar and 255)

    # Salvar a imagem em preto e branco
    nome_arquivo_pb = nome_arquivo.replace('.jpg', 'pb.jpg')
    imagem_pb.save(nome_arquivo_pb)

    print(f"Imagem em escala de cinza salva como '{nome_arquivo_cinza}'")
    print(f"Imagem em preto e branco salva como '{nome_arquivo_pb}'")

# Chamar a função com o nome do arquivo da imagem
converter_imagem('dinamite10.jpg')
```

```python
import matplotlib.pyplot as plt

def exibir_imagens(nome_original, nome_cinza, nome_pb):
    # Carregar e exibir a imagem original
    plt.subplot(1, 3, 1)
    imagem_original = plt.imread(nome_original)
    plt.imshow(imagem_original)
    plt.title('Original')
    plt.axis('off')

    # Carregar e exibir a imagem em escala de cinza
    plt.subplot(1, 3, 2)
    imagem_cinza = plt.imread(nome_cinza)
    plt.imshow(imagem_cinza, cmap='gray')
    plt.title('Escala de Cinza')
    plt.axis('off')

    # Carregar e exibir a imagem em preto e branco
    plt.subplot(1, 3, 3)
    imagem_pb = plt.imread(nome_pb)
    plt.imshow(imagem_pb, cmap='gray')
    plt.title('Preto e Branco')
    plt.axis('off')

    # Exibir as imagens
    plt.show()

# Chamar a função para exibir as imagens
exibir_imagens('dinamite10.jpg', 'dinamite10cinza.jpg', 'dinamite10pb.jpg')
```
